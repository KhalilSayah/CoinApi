```python
list_token = ["CVC","RARI","AAVE","UNI","ATOM","THETA","SAND","MAGIC","TWT","AMP"]
```

```python
def get_top_symbol_id(df, token, top):
    # Filter the DataFrame for the specified token and ensure data_orderbook_start is not NaN
    f = df[(df["asset_id_base"] == token) & (df["data_orderbook_start"].notna())]

    # Sort the filtered DataFrame by volume_1day_usd in descending order
    f = f.sort_values(by="volume_1mth_usd", ascending=False)

    # Select the top rows based on the specified number
    r = f[["symbol_id", "volume_1mth_usd", "price", "data_orderbook_start"]].head(top)

    return r
```

```python
def get_top_symbols_for_tokens(df, tokens, top):
    results = []
    for token in tokens:
        top_symbols = get_top_symbol_id(df, token, top)
        top_symbols['asset_id_base'] = token  # Add token as a new column for identification
        results.append(top_symbols)  # Append the results to the list

    # Concatenate all results into a single DataFrame
    aggregated_results = pd.concat(results, ignore_index=True)
    return aggregated_results
```

```python
focus_symbols = get_top_symbols_for_tokens(df,list_token, 5)
```

```python
focus_symbols.to_json("focus_symbols.json")
```
